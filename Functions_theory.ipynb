2. Explain the concept of function arguments and parameters in Python. 

    Parameters are variables defined within the parentheses of a function's def statement. They act as placeholders for the values that the function expects to receive when it is called.

    Arguments are the actual values passed to a function when it is called. These values are assigned to the corresponding parameters within the function's definition.
